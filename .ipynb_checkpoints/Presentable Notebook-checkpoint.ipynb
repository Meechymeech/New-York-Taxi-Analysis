## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib
```

## Import Data

```python
taxi_jan_2025 = pd.read_parquet('data/yellow_tripdata_2025-01.parquet')
taxi_data = pd.concat([taxi_jan_2025])
```

```python
taxi_data.head()
```

```python
taxi_data.shape
```

```python
taxi_data.info()
```

```python
taxi_data = taxi_data.drop(columns=['store_and_fwd_flag','fare_amount','extra','mta_tax','tip_amount','tolls_amount',
                                    'improvement_surcharge','congestion_surcharge','Airport_fee'])
```

```python
taxi_data = taxi_data.drop(columns=['VendorID'])
taxi_data.head()
```
